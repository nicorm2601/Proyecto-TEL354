```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Función para arreglar la primera columna del archivo csv.
def eliminarColumna(ruta_dataset):
    df = pd.read_csv(ruta_dataset)
    df = df.drop(df.columns[0], axis=1)
    df.to_csv(ruta_dataset, index=False)

    return ruta_dataset
```

```python
archivo_ordenado = eliminarColumna('camila_vallejo_sent.csv')
```

```python
def visualizar_mes(ruta_in_csv, user):
    df = pd.read_csv(ruta_in_csv)
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.to_period('M')
    sentiment_by_month = df.groupby(['month', 'sentiment']).size().unstack().fillna(0)

    sns.set_theme(style = "whitegrid")

    sentiment_by_month.plot(kind = 'bar', stacked = True, figsize = (12, 6))

    plt.xlabel('Mes')
    plt.ylabel('Cantidad de Tweets')
    plt.title("Análisis de Sentimientos de " + user)
    plt.legend(title = 'Sentimientos')
    plt.xticks(rotation = 45)
    plt.tight_layout()

    plt.show()
```

```python
visualizar_mes('joseantoniokast_sent.csv', 'Jose Antonio Kast')
```

```python
visualizar_mes('GabrielBoric_sent.csv', 'Gabriel Boric')
```

```python
visualizar_mes('camila_vallejo_sent.csv', 'Camila Vallejo')
```
