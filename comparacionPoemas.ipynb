```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer los archivos csv de los autores analizados.
df_autor1 = pd.read_csv('./Poemas/Antonio_Machado_sent.csv')
df_autor2 = pd.read_csv('./Poemas/Gabriela_Mistral_sent.csv')
df_autor3 = pd.read_csv('./Poemas/Mario_Benedetti_sent.csv')
df_autor4 = pd.read_csv('./Poemas/Nicanor_Parra_sent.csv')
df_autor5 = pd.read_csv('./Poemas/Pablo_Neruda_sent.csv')
df_autor6 = pd.read_csv('./Poemas/Vicente_Huidobro_sent.csv')

df = pd.concat([df_autor1, df_autor2, df_autor3, df_autor4, df_autor5, df_autor6])
```

```python
# Función para crear el grafico de comparación de emciones en los poemas.
def comparacionEmociones(dataframe):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(14, 7))
    sns.countplot(data=dataframe, x='predicted_emotion', hue='author')
    plt.title('Comparación de Emociones Predichas por Autor')
    plt.xlabel('Emociones Predichas')
    plt.ylabel('Número de Poemas')
    plt.legend(title='Autor')
    plt.xticks(rotation=45)
    
    plt.show()
```

```python
# Función para crear el grafico de comparación de sentimientos en los poemas.
def comparacionSentimientos(dataframe):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(14, 7))
    sns.countplot(data=dataframe, x='predicted_sentiment', hue='author')
    plt.title('Comparación de Sentimientos Predichos por Autor')
    plt.xlabel('Sentimientos Predichos')
    plt.ylabel('Número de Poemas')
    plt.legend(title='Autor')
    plt.xticks(rotation=45)
    plt.show()
        
    plt.show()
```

```python
comparacionEmociones(df)
```

```python
comparacionSentimientos(df)
```
