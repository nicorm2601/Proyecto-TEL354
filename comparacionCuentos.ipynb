```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer los archivos csv de los autores analizados.
df_cuento1 = pd.read_csv('./Cuentos/Alfonso_Reyes_sent.csv')
df_cuento2 = pd.read_csv('./Cuentos/Baldomero_Lillo_sent.csv')
df_cuento3 = pd.read_csv('./Cuentos/Jorge_Luis_Borges_sent.csv')
df_cuento4 = pd.read_csv('./Cuentos/Julio_Cortazar_sent.csv')
df_cuento5 = pd.read_csv('./Cuentos/Maria_Luisa_Bombal_sent.csv')
df_cuento6 = pd.read_csv('./Cuentos/Mario_Benedetti_sent.csv')
df_cuento7 = pd.read_csv('./Cuentos/Vicente_Huidobro_sent.csv')

df = pd.concat([df_cuento1, df_cuento2, df_cuento3,df_cuento4, df_cuento5, df_cuento6, df_cuento7])
```

```python
# Función para crear el grafico de comparación de emociones en los cuentos.
def comparacionEmociones(dataframe):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(14, 7))
    sns.countplot(data=dataframe, x='predicted_emotion', hue='author')
    plt.title('Comparación de Emociones Predichas por Autor')
    plt.xlabel('Emociones Predichas')
    plt.ylabel('Número de Poemas')
    plt.legend(title='Autor')
    plt.xticks(rotation=45)
    
    plt.show()
```

```python
# Función para crear el grafico de comparación de sentimientos en los cuentos.
def comparacionSentimientos(dataframe):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(14, 7))
    sns.countplot(data=dataframe, x='predicted_sentiment', hue='author')
    plt.title('Comparación de Sentimientos Predichos por Autor')
    plt.xlabel('Sentimientos Predichos')
    plt.ylabel('Número de Poemas')
    plt.legend(title='Autor')
    plt.xticks(rotation=45)
    plt.show()
        
    plt.show()
```

```python
comparacionEmociones(df)
```

```python
comparacionSentimientos(df)
```
