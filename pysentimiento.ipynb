```python
from pysentimiento import create_analyzer
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("camila_vallejo_org.csv")
```

```python
df
```

```python
# Analizar la emoción.

analyzer = create_analyzer(task="emotion", lang="es")

df = df.dropna(subset=['text'])

def get_main_emotion(text):
    result = analyzer.predict(text)
    main_emotion = max(result.probas, key=result.probas.get)
    
    return main_emotion

df['emotion'] = df['text'].apply(get_main_emotion)
```

```python
df
```

```python
# Analizar el sentimiento.

analyzer2 = create_analyzer(task="sentiment", lang="es")

df = df.dropna(subset=['text'])

def get_main_sentiment(text):
    result = analyzer2.predict(text)
    main_sentiment = max(result.probas, key=result.probas.get)
    
    return main_sentiment

df['sentiment'] = df['text'].apply(get_main_sentiment)
```

```python
df
```

```python
df.to_csv('camila_vallejo_sent.csv', header=True)
```
